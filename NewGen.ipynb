## Desarrollo de la solución

Carga y Visualización Inicial de los Datos: El primer paso consiste en obtener los datos históricos de las acciones de Amazon utilizando la API de Yahoo Finance. Se descargaron datos con un intervalo de 15 minutos para un periodo de 10 días, lo que proporciona una buena granularidad para el análisis.

## Análisis de Amazon

```python
import yfinance as yf
from datetime import date, timedelta

# Test dates
today = date.today() 
back58days = today - timedelta(days=10)
dataF = yf.download("AMZN", start=back58days, end=today, interval="15m")
```

```python
dataF.head(10)
```

Los datos incluyen las siguientes columnas:

Open: Precio de apertura de la acción.

High: Precio máximo alcanzado durante el intervalo.

Low: Precio mínimo alcanzado durante el intervalo.

Close: Precio de cierre de la acción.

Volume: Número de acciones negociadas.

```python
dataF.describe()
```

```python
dataF.info()
```

```python
## Visualización de los datos:

import matplotlib.pyplot as plt

# Gráfico de la evolución del precio de cierre
plt.figure(figsize=(10, 6))
plt.plot(dataF['Close'], label='Precio de Cierre')
plt.title('Evolución del Precio de Cierre de Amazon')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.legend()
plt.show()
```

```python
# Gráfico del volumen negociado
plt.figure(figsize=(10, 6))
plt.bar(dataF.index, dataF['Volume'], label='Volumen Negociado')
plt.title('Volumen Negociado de Amazon')
plt.xlabel('Fecha')
plt.ylabel('Volumen')
plt.show()
```

```python
## cálculo de medias móviles
dataF['SMA_50'] = dataF['Close'].rolling(window=50).mean()

# Gráfico de las medias móviles
plt.figure(figsize=(10, 6))
plt.plot(dataF['Close'], label='Precio de Cierre')
plt.plot(dataF['SMA_50'], label='Media Móvil 50 Periodos', linestyle='--')
plt.title('Tendencias con Medias Móviles')
plt.xlabel('Fecha')
plt.ylabel('Precio (USD)')
plt.legend()
plt.show()
```

### Explicación: 

Precio de Cierre (línea azul):

La línea azul representa el precio de cierre de las acciones de Amazon en intervalos de 15 minutos. Podemos observar la volatilidad del mercado, con subidas y bajadas bruscas en este periodo de tiempo.
En la primera parte de la gráfica, el precio muestra una tendencia ascendente, pero luego comienza a fluctuar más drásticamente con caídas pronunciadas, especialmente hacia el final de los días representados.


Media Móvil de 50 Periodos (línea naranja discontinua):

La media móvil de 50 periodos es una medida suavizada que promedia los precios de los últimos 50 periodos (en este caso, intervalos de 15 minutos). Su función es destacar la tendencia a corto plazo de la serie de precios, eliminando el ruido de las fluctuaciones momentáneas.
En la gráfica, la media móvil de 50 periodos se mantiene relativamente estable y sigue de cerca las fluctuaciones del precio de cierre, aunque con menos volatilidad. Sin embargo, durante las caídas pronunciadas en el precio, la media móvil responde con cierta demora debido a su naturaleza suavizada

#### Razón de la Forma de la Gráfica:


Volatilidad: La naturaleza fluctuante del precio de cierre en la línea azul indica una alta volatilidad en las acciones de Amazon durante el periodo analizado. Este tipo de comportamiento es común en mercados financieros, especialmente en el trading de alta frecuencia.


Diferencia entre las medias móviles y el precio de cierre: Las medias móviles reaccionan más lentamente a los cambios abruptos en el precio de cierre debido a que promedian los precios de un periodo más largo. Es por esto que, aunque el precio de cierre cae rápidamente, la media móvil de 50 periodos sigue una trayectoria más estable y con menos cambios abruptos.

## EDA sobre Google

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

google_stock = yf.download('GOOGL', start='2015-01-01', end='2024-01-01')
```

```python
google_stock.head()
```

```python
google_stock.info()
```

```python
google_stock.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(google_stock['Adj Close'], label='Precio de Cierre Ajustado', color='blue')
plt.title('Precio de Cierre Ajustado de Google')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.show()
```

Esta simple gráfica es muy útil, nos da la información de que el precio de la acción de Google tiene a ir a la alza desde hace 9 años y podemos observar cierta estacionalidad relacionada con los meses del año, aunque lo más remarcable es que las caidas tienden a ser abruptas y el crecimiento mas controlado.

```python
# Calcular los retornos diarios
google_stock['Daily Return'] = google_stock['Adj Close'].pct_change()

plt.figure(figsize=(10, 6))
sns.histplot(google_stock['Daily Return'].dropna(), bins=50, color='purple', kde=True)
plt.title('Distribución de los Retornos Diarios de Google')
plt.xlabel('Retorno Diario')
plt.ylabel('Frecuencia')
plt.show()
```

Podemos observar que la mayoría de dias el valor de retorno diario de la acción está en torno a cero, lo que nos quiere decir que esta acción tiene un precio bastante estable, pero existen los valores mas extremos que nos indican que hay días en los que la acción cambia de precio mas bruscamente.

```python
# vemos los valores mas bajos de retornos diarios
top_negative_returns = google_stock['Daily Return'].nsmallest(5)
print("\nValores más bajos (negativos):")
print(top_negative_returns)
```

```python
google_stock['Month'] = google_stock.index.month

plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Adj Close', data=google_stock)
plt.title('Distribución del Precio de Cierre Ajustado por Mes')
plt.xlabel('Mes')
plt.ylabel('Precio de Cierre Ajustado')
plt.show()
```

Acá podemos ver que los precios tienden a ser un poco más altos en verano y a bajar en el final e inicio del año. Pero hay outliers altos en todos los meses, excpetuando mayo, junio y julio.
